```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/devensinghbhagtani/Bollywood-Movie-Dataset/refs/heads/main/IMDB-Movie-Dataset(2023-1951).csv")
```

```python
df=pd.DataFrame(df)
```

```python
df
```

```python
df["tags"]=df["genre"]+ " " + df["overview"]+ " " + df["director"]+ " " + df["cast"]
```

```python
final = df[['movie_id','movie_name',"tags"]]
```

```python
final
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv
```

```python
vector=cv.fit_transform(final["tags"])
```

```python
vector.toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
final["tags"] = final["tags"].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie_name):

    if movie_name not in final["movie_name"].values:
        print("Movie not found in database ❌")
        return
    movie_index = final[final["movie_name"] == movie_name].index[0]

    distances = similarity[movie_index]
    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    print(f"\nTop 5 movies similar to '{movie_name}':\n")

    for i in movies_list:
        print(final["movie_name"].iloc[i[0]])
```

```python
recommend("Kyo Kii... Main Jhuth Nahin Bolta")
```

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(final, f)

with open("similarity.pkl", "wb") as f:
    pickle.dump(similarity, f)
```
